```python
import pandas as pd

# Load the dataset
data_path = 'reviews.csv'
data = pd.read_csv(data_path)

# Replace non-numeric values with NaN
data['Overall_Rating'] = pd.to_numeric(data['Overall_Rating'], errors='coerce')

# Display the unique values in the 'Overall_Rating' column after replacement
data['Overall_Rating'].unique()
```

```python
# Fill missing values in the "Overall_Rating" column with the median value
data['Overall_Rating'].fillna(data['Overall_Rating'].median(), inplace=True)

# Define the bins and labels for the 'Rating_Class' column
bins = [0, 2.5, 3.5, 4.5, 5]
labels = ['Poor', 'Average', 'Good', 'Excellent']

# Create the 'Rating_Class' column
data['Rating_Class'] = pd.cut(data['Overall_Rating'], bins=bins, labels=labels, include_lowest=True)

# Display the first few rows of the dataframe
data.head()
```
